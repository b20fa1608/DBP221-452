```python
import pandas as pd
import numpy as np
```

```python
### Оюутан уг файлыг Python-д оруулан ашиглана эсвэл уг файлын датаг sheet тус бүрээр үүгэнэ.
```

```python
irts = pd.read_excel("C:/Users/Dell/OneDrive/Documents/2-р курс/2-2/Бие даалт - 1.xlsx", sheet_name = 'Ирц')
irts = irts.dropna(axis = 1, how = 'all')
irts = irts.fillna(0)
irts.replace(('и','т'),(0.3125,0), inplace = True)
del irts['Нийт -10']
del irts['Unnamed: 37']
a = irts.T[3:35]
sum_irts = a.sum(axis=0)
irts['Нийт'] = sum_irts
irts
```

```python
seminar = pd.read_excel("C:/Users/Dell/OneDrive/Documents/2-р курс/2-2/Бие даалт - 1.xlsx", sheet_name = 'Семинар')
seminar = seminar.dropna(axis = 1, how = 'all')
seminar = seminar.fillna(0)
del seminar['Нийт -25']
b = seminar.T[3:35]
sum_seminar = b.sum(axis=0)
seminar['Нийт'] = sum_seminar
seminar
```

```python
BiyDaalt= pd.read_excel("C:/Users/Dell/OneDrive/Documents/2-р курс/2-2/Бие даалт - 1.xlsx", sheet_name = 'Бие даалт')
BiyDaalt = BiyDaalt.dropna(axis = 1, how = 'all')
BiyDaalt = BiyDaalt.fillna(0)
del BiyDaalt['Нийт']
c = BiyDaalt.T[3:35]
sum_BiyDaalt = c.sum(axis=0)
BiyDaalt['Нийт'] = sum_BiyDaalt
BiyDaalt
```

```python
shalgalt = pd.read_excel("C:/Users/Dell/OneDrive/Documents/2-р курс/2-2/Бие даалт - 1.xlsx", sheet_name = 'Шалгалт')
shalgalt = shalgalt.dropna(axis = 1, how = 'all')
shalgalt = shalgalt.fillna(0)
del shalgalt['Нийт']
d = shalgalt.T[3:35]
sum_shalgalt = d.sum(axis=0)
shalgalt['Нийт'] = sum_shalgalt
shalgalt
```

```python
humuujil_handlaga = pd.read_excel("C:/Users/Dell/OneDrive/Documents/2-р курс/2-2/Бие даалт - 1.xlsx", sheet_name = 'Хүмүүжил хандлага')
humuujil_handlaga = humuujil_handlaga.dropna(axis = 1, how = 'all')
humuujil_handlaga = humuujil_handlaga.fillna(0)
del humuujil_handlaga['Нийт']
e = humuujil_handlaga.T[3:35]
sum_humuujil_handlaga = e.sum(axis=0)
humuujil_handlaga['Нийт'] = sum_humuujil_handlaga
humuujil_handlaga
```

```python
nemelt_onoo = pd.read_excel("C:/Users/Dell/OneDrive/Documents/2-р курс/2-2/Бие даалт - 1.xlsx", sheet_name = 'Нэмэлт оноо')
nemelt_onoo = nemelt_onoo.dropna(axis = 1, how = 'all')
nemelt_onoo = nemelt_onoo.fillna(0)
del nemelt_onoo['Нийт']
f = nemelt_onoo.T[3:35]
sum_nemelt_onoo = f.sum(axis=0)
nemelt_onoo['Нийт'] = sum_nemelt_onoo
nemelt_onoo
```

```python
x = nemelt_onoo['Оюутны код']
y = nemelt_onoo['Оюутны нэр']
```

```python
Niilber = pd.DataFrame({'Оюутны код':x,'Оюутны нэр':y,'Ирц':sum_irts,'Семинар':sum_seminar,'Шалгалт':sum_shalgalt,'Хүмүүжил Хандлага':sum_humuujil_handlaga,'Бие даалт':sum_BiyDaalt,'Нэмэлт Оноо':sum_nemelt_onoo})
Niilber['Нийт'] = Niilber.sum(axis=0)
h = Niilber.T[2:8].sum(axis=0)
Niilber = pd.DataFrame({'Оюутны код':x,'Оюутны нэр':y,'Ирц':sum_irts,'Семинар':sum_seminar,'Шалгалт':sum_shalgalt,'Хүмүүжил Хандлага':sum_humuujil_handlaga,'Бие даалт':sum_BiyDaalt,'Нэмэлт Оноо':sum_nemelt_onoo,'Нийт':h})
Niilber
```

```python
grade = []
for i in h:
    if i>=95.5 and i<=100:
        grade.append("A")
    elif i>=89.5 and i<95.5:
        grade.append("A-")
    elif i>=86.5 and i<89.5:
        grade.append("B+")
    elif i>=82.5 and i<86.5:
        grade.append("B")
    elif i>=79.5 and i<82.5:
        grade.append("B-")
    elif i>=76.5 and i<79.5:
        grade.append("C+")
    elif i>=72.5 and i<76.5:
        grade.append("C")
    elif i>=69.5 and i<72.5:
        grade.append("C-")
    elif i>=66.5 and i<69.5:
        grade.append("D+")
    elif i>=62.5 and i<66.5:
        grade.append("D")
    elif i>=59.5 and i<62.5:
        grade.append("D-")
    elif i>=0 and i<59.5:
        grade.append("F")
else:
        print("")

Niilber['Үнэлгээ'] = grade
Niilber

```

```python
davtamj = Niilber.pivot_table(columns=['Үнэлгээ'], aggfunc = 'size')
print(davtamj)
dugnelt = pd.DataFrame(davtamj)
```

```python
print('Үнэлгээний хураангуй:',dugnelt)
dugnelt
```

```python
amjilt = pd.DataFrame({'Тэнцсэн хүүхдийн тоо': dugnelt[0:8].sum()/len(Niilber)*100})
print(amjilt)
```

```python
surgaltiin_chanar = pd.DataFrame({'Сургалтын чанар':dugnelt[0:4].sum()/len(Niilber)*100})
```
